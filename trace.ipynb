```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataDict = {
'State' : pd.Series(['FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM'],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
'Up' : pd.Series([12.58797852,7.780854465,12.51398955,7.682605196,12.23329953,7.712529515,12.18530505,7.724823197,12.147342,7.600728557,12.03833525,7.631163288,11.97165255,7.505025956],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
'Up Error': pd.Series([],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
'Down' : pd.Series([3.108956282,7.764905323,2.925061091,7.795722144,3.043685565,7.700277302,3.05377276,7.649221414,3.07346697,7.642687452,2.804565129,7.68348004,2.906474074,7.53305786],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90'])
'Down Error': pd.Series([],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
}
```

```python
dataDict = pd.DataFrame(dataDict)
dataDict
```

```python
angle =[0,15,30,45,60,75,90]
upafm = []
upfm = []

for index, row in dataDict.iterrows():
    if row[1] == 'FM':
        upfm.append(row[2])
    else:
        upafm.append(row[2])
        
upafm = np.asarray(upafm)
upfm = np.asarray(upfm)
```

```python
downafm = []
downfm = []

for index, row in dataDict.iterrows():
    if row[1] == 'FM':
        downfm.append(row[0])
    else:
        downafm.append(row[0])
        
downafm = np.asarray(downafm)
downfm = np.asarray(downfm)
```

```python
plt.plot(angle,upafm,marker='o',label="Up")
plt.plot(angle,downafm,marker='o',label="Down")
plt.ylabel('Trace AFM')
plt.xlabel('Angle')
plt.title('Trace Up and Down AFM vs. Angle')
plt.xlim(-5, 100)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(angle,downafm,marker='o',label="Down")
plt.ylabel('Trace Down AFM')
plt.xlabel('Angle')
plt.title('Trace Down AFM vs. Angle')
plt.xlim(-5, 100)
plt.show()
```
