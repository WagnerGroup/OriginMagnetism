```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataDict = {
'State' : pd.Series(['FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM','FM','AFM'],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
'Up' : pd.Series([11.57933117,7.725928567,11.52673443,7.725928567,11.31191627,7.689469784,11.17892921,7.684781534,11.14873758,7.56531651,11.06889553,7.617588276,11.05401123,7.513606659],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90']),
'Down' : pd.Series([2.98404325,6.756112109,2.949075622,6.756112109,2.936140546,6.687459851,2.994861386,6.632153174,3.009565377,6.658311012,2.898540869,6.653985282,2.94586064,6.552924104],index=['0','0', '15','15', '30','30','45','45','60','60','75','75','90', '90'])
}
```

```python
dataDict = pd.DataFrame(dataDict)
dataDict
```

```python
angle =[0,15,30,45,60,75,90]
upafm = []
upfm = []

for index, row in dataDict.iterrows():
    if row[1] == 'FM':
        upfm.append(row[2])
    else:
        upafm.append(row[2])
        
upafm = np.asarray(upafm)
upfm = np.asarray(upfm)
```

```python
downafm = []
downfm = []

for index, row in dataDict.iterrows():
    if row[1] == 'FM':
        downfm.append(row[0])
    else:
        downafm.append(row[0])
        
downafm = np.asarray(downafm)
downfm = np.asarray(downfm)
```

```python
plt.plot(angle,upafm,marker='o',label="Up")
plt.plot(angle,downafm,marker='o',label="Down")
plt.ylabel('Trace AFM')
plt.xlabel('Angle')
plt.title('Trace Up and Down AFM vs. Angle')
plt.xlim(-5, 100)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(angle,downafm,marker='o',label="Down")
plt.ylabel('Trace Down AFM')
plt.xlabel('Angle')
plt.title('Trace Down AFM vs. Angle')
plt.xlim(-5, 100)
plt.show()
```
