```python
dataDict = pd.DataFrame(dataDict)
dataDict
```

```python
angle =[0,15,30,45,60,75,90]
upafm = []
upfm = []
uperrafm=[]
uperrfm = []

for index, row in dataDict.iterrows():
    if row[2] == 'FM':
        upfm.append(row[3])
        uperrfm.append(row[4])
    else:
        upafm.append(row[3])
        uperrafm.append(row[4])
upafm = np.asarray(upafm)
upfm = np.asarray(upfm)
uperrafm=np.asarray(uperrafm)
uperrfm=np.asarray(uperrfm)
```

```python
downafm = []
downfm = []
downerrafm=[]
downerrfm = []
for index, row in dataDict.iterrows():
    if row[2] == 'FM':
        downfm.append(row[0])
        downerrfm.append(row[1])
    else:
        downafm.append(row[0])
        downerrafm.append(row[1])
        
downafm = np.asarray(downafm)
downfm = np.asarray(downfm)
downerrafm=np.asarray(downerrafm)
downerrfm=np.asarray(downerrfm)
```

```python
plt.errorbar(angle,upafm,yerr=uperrafm,marker='o',label="Up")
plt.errorbar(angle,downafm,yerr=downerrafm,marker='o',label="Down")
plt.ylabel('Trace AFM')
plt.xlabel('Angle')
plt.title('Trace Up and Down AFM vs. Angle')
plt.xlim(-5, 100)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
